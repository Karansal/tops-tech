```python
# 4. In a standardized test, scores follow a normal distribution with mean = 100, std. dev = 15:
# What is the probability that a student scores above 115?
# What percentage of students scored between 85 and 115?

from scipy.stats import norm
```

```python
mean = 100
std_dev = 15
```

```python
p_above_115 = 1 - norm.cdf(115, mean, std_dev)
```

```python
p_between_85_115 = norm.cdf(115, mean, std_dev) - norm.cdf(85, mean, std_dev)
```

```python
percent_above_115 = p_above_115 * 100
percent_between_85_115 = p_between_85_115 * 100
```

```python
print(f"Probability (score > 115): {p_above_115:.4f}  -->  {percent_above_115:.2f}%")
```

```python
print(f"Probability (85 <= score <= 115): {p_between_85_115:.4f}  -->  {percent_between_85_115:.2f}%")
```

Create a histogram and verify the data follows a normal distribution.
Calculate Z-score for a student who scored 130.
Interpret the Z-score meaning in this context.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores = [37,23,59,54,44,27,24,55,56,25,36,53,24,37,25,34,12,20,14,27,37,31,
14,49,22,31,35,29,17,19,54,44,47,36,22,50,12,59,22,48,52,19,29,42,36,13,27,
19,15,42,24,45,17,33,12,20,50,51,37,46,42,41,57,49,17,30,47,33,40,18,14,55,
33,31,17,32,57,47,48,58,42,48,35,54,53,17,39,58,27,21,57,38,57,58,49,39,49,
20,41,20]
```

```python
mean = np.mean(scores)
std = np.std(scores)
```

```python
print("Mean:", round(mean, 2))
print("Standard Deviation:", round(std, 2))
```

```python
plt.hist(scores, bins=10, density=True, alpha=0.6, color='skyblue', edgecolor='black')
```

```python
z_130 = (130 - mean) / std
print("Z-score for a score of 130:", round(z_130, 2))
```

```python
if z_130 > 3:
    interpretation = "This score is extremely high and very unusual compared to others."
elif z_130 > 2:
    interpretation = "This score is much higher than average â€” exceptional performance."
elif z_130 > 1:
    interpretation = "This score is above average."
else:
    interpretation = "This score is within the normal range."

print("Interpretation:", interpretation)
```
